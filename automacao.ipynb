## Problema a ser resolvido
- buscar automaticamente os dados de ações no site do yahoo finance
- analisar os dados de ações de acordo com a solicitação do gestor
- enviar de forma automática um e-mail com todas as analises

```python
import yfinance, pyautogui, pyperclip
from time import sleep
```

## Passo 1 - Buscar automaticamente os dados das ações no site

```python
codigo_acao = input("Digite o código da ação: ")
dados_acao = yfinance.Ticker(codigo_acao).history("6mo")

fechamento_acao = dados_acao.Close

fechamento_acao.plot()
```

## Passo 2  - Analises solicitadas pelo gestor
- Analise dos últimos seis meses
- Cotação Máxima
- Cotação Mínima
- Cotação Atual


```python
#últimos seis meses
grafico_acao = fechamento_acao.plot()
#cotação máxima
cotacao_maxima = fechamento_acao.max()
#cotação mínima
cotacao_minima = fechamento_acao.min()
#cotação atual
cotacao_atual = fechamento_acao.iloc[-1]
```

## Automatizando o envio do E-mail

```python
pyautogui.PAUSE = 1.5

pyautogui.press('win')
pyautogui.write('google')
pyautogui.press('enter')

pyperclip.copy('www.gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('enter')

pyautogui.click(x=80, y=184)

pyperclip.copy('google@gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

pyperclip.copy('Análise Diária de Ação')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

mensagem_corpo = f'''
Prezado <Pessoa>,

Segue as análises da Ação {codigo_acao}:

Cotação Máxima: R$ {round(cotacao_maxima, 2)}
Cotação Mínima: R$ {round(cotacao_minima, 2)}
Cotação Atual: R$ {round(cotacao_atual, 2)}

Obs.: Este e-mail não deve ser respondido, pois foi enviado por um bot de aprendizado meu.

Até mais.
'''
pyperclip.copy(mensagem_corpo)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

pyautogui.press('enter')
```
